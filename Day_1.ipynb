### Solution :

```python
# Function to check if the average of A and B is greater than C
def is_greater_average(A, B, C):
    average = (A + B) / 2
    return average > C

# Read the number of test cases
T = int(input())

# Iterate through each test case
for _ in range(T):
    A, B, C = map(int, input().split())
    if is_greater_average(A, B, C):
        print("YES")
    else:
        print("NO")
```
